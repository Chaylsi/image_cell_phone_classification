```python
from google.colab import drive
file1 = ("/content/drive/MyDrive/chancedz/training/training")
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(validation_split=0.33, horizontal_flip=True)

train_set = train_datagen.flow_from_directory(file1,
                                              target_size = (384,384),
                                              batch_size = 4,
                                              subset = 'training',
                                              class_mode='binary')

import tensorflow as tf

test_set = train_datagen.flow_from_directory(file1,
                                             target_size = (384,384),
                                             batch_size = 2,
                                             subset = 'validation',
                                             class_mode='binary')
```

```python
cnn = tf.keras.models.Sequential()
cnn.add(tf.keras.layers.Dense(1, activation= 'relu',input_shape=[384,384, 3]))
cnn.add(tf.keras.layers.Conv2D(128, kernel_size=[5,5], padding='valid', activation='relu'))
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=[4,4] , strides=2, padding='valid'))
cnn.add(tf.keras.layers.Conv2D(64, kernel_size=[3,3], padding='valid', activation='relu'))
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=[3,3] , strides=2, padding='valid'))
cnn.add(tf.keras.layers.Conv2D(32, kernel_size=[2,2], padding='valid', activation='relu'))
cnn.add(tf.keras.layers.Flatten())
cnn.add(tf.keras.layers.Dense(1, activation= 'sigmoid'))
cnn.compile(optimizer= 'adam' , loss ='binary_crossentropy', metrics=['accuracy'])
cnn.summary()

history = cnn.fit(train_set, validation_data =test_set, epochs=13, verbose=2)
```

```python
cnn.save("sell_pred.h5")
```

```python
import cv2
def prepare(image):
  IMG_SIZE=384

  img_array = cv2.imread('/content/drive/MyDrive/chancedz/test_simp/test_simp/20.jpg',  cv2.IMREAD_COLOR)
  new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
  return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 3)

x=cnn.predict([prepare('/content/drive/MyDrive/chancedz/test_simp/test_simp/20.jpg')])

print(x)
score = x[0]
print("this is %.2f percent chance no cell phone and %.2f cell phone."
%(100*(1-score),100*score)
)
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/chancedz/test_simp/test_simp/20.jpg',0)
plt.imshow(img, interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```
